```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import scipy
%matplotlib inline
```

```python
def function(x):
    
    y = 16*(x**4) - 32*(x**3) - 8*(x**2) + 10*x + 9
    
    return y
```

```python
ls = []
iter = np.arange(-2.0, 3.0, 0.1)
for i in iter:
    y = function(i)
    ls.append(y)
    
plt.figure(figsize = (12, 8))
plt.plot(iter, ls)
plt.title("Function")
```

```python
# Derivative of the equation comes out to be: 32*(x**3) - 48*(x**2) -8*x + 5

### Solving the cubic equation using Numpy
roots = np.roots([32, -48, -8, 5])
print "Global minima of the function is at x =", roots[0]
print "Local minima of the function is at x =", roots[1]
```

## Running Gradient Descent at given values

```python
def derivative_function(x):
    
    y =  64*(x**3) - 96*(x**2) - 16*x + 10
    
    return y
```

```python
# Gradient Descent algorithm

def gradient_descent(x, eta, iterations):

    grad_ls = []

    for i in range(0, iterations):
        
        x = x - eta*(derivative_function(x))
        grad_ls.append(x)

    return grad_ls
```

```python
# Running Gradient Descent for 1000 iterations and plotting the results

def plot_grad_descent(x, eta, iterations):
    
    grad_ls = gradient_descent(x, eta, iterations)
    print "First 5 values of x:", grad_ls[:5]

    fig1 = plt.figure(figsize=(12, 8))
    fig1 = plt.plot(grad_ls[:5])
    fig1 = plt.title("First 5 iterations")

    print "Last 5 values of x:", grad_ls[-5:]

    fig2 = plt.figure(figsize=(12, 8))
    fig2 = plt.plot(grad_ls)
    fig2 = plt.title("1000 iterations")
    
    # Value of f(x) using updated weights
    fx_ls = []
    for i in grad_ls:
        y = function(i)
        fx_ls.append(y)

    print "First 5 values of f(x):", fx_ls[:5]
    print "Last 5 values of f(x):", fx_ls[-5:]
```

### (b) x = -1, eta = 0.001

```python
plot_grad_descent(-1, 0.001, 1000)
```

### (c)  x = 2, eta = 0.001

```python
plot_grad_descent(2, 0.001, 1000)
```

### (d) x = -1, eta = 0.01

```python
# Running Gradient Descent at a higher learning rate will help us bounce off from local minima

plot_grad_descent(-1, 0.01, 1000)
```
